```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define the range of input sizes
n = np.linspace(1, 100, 400)  # 1 to 100, with 400 points

# Define the functions for different Big O notations
O_1 = np.ones_like(n)  # Constant time complexity
O_log_n = np.log2(n)  # Logarithmic time complexity
O_n = n               # Linear time complexity
O_n_log_n = n * np.log2(n)  # Linearithmic time complexity
O_n_squared = n**2    # Quadratic time complexity

# Plotting
plt.figure(figsize=(10, 6))

plt.plot(n, O_1, label='O(1)', color='blue')
plt.plot(n, O_log_n, label='O(log n)', color='green')
plt.plot(n, O_n, label='O(n)', color='red')
plt.plot(n, O_n_log_n, label='O(n log n)', color='purple')
plt.plot(n, O_n_squared, label='O(n^2)', color='orange')

plt.ylim(0, 10000)  # Set y-axis limit to ensure all plots are visible
plt.xlim(0, 100)    # Set x-axis limit to ensure all plots are visible

plt.xlabel('Input Size (n)')
plt.ylabel('Operations')
plt.title('Big O Notation of Different Run Times')
plt.legend()
plt.grid(True)
plt.savefig('big_o_notation_plot.png')
# Show the plot
plt.show()
```
